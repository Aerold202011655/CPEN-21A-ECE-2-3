##For Loop

```python
week=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
```

##Break Statement

```python
week=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
  if x=="Thursday":
    break
```

```python
for x in week:
  if x=="Thursday":
    break
  print(x)
```

##Looping through a string

```python
for x in "Programming Logic and Design":
  print(x)
```

##The Range Function

```python
for x in range(6):
  print(x)
```

```python
for x in range(2,6):
  print(x)
```

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple","banana","cherry"]

for x in adjective:
  for y in fruits:
    print(x,y)
```

##While Loop

```python
i = 1
while i<=6:
  print(i)
  i+=1  #The same as i=i+1#Assignment operator for addition
```

The break statement

```python
i=1
while i<6:
  print(i)
  if i==3:
    break
  else:
   i+=1
```

The continued statement

```python
i=1

while i<6:
  print(i)
  if i==3:
    continue
  i+=1
```

The else statement

```python
i=1
while i<6:
  print(i)
  i+=1
else:
  print("i is no longer less than 6")
```
